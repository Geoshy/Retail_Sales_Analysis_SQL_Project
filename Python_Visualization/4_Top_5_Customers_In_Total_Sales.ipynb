# **Top 5 Customers According To Sum Of The Total Sales:**

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("D:\\IT Courses\\Data Analysis Courses\\Data Analysis Projects\\Retail Sales Analysis SQL Project\\Retail_Sales_Analysis_SQL_Project\\CSV_Files\\problem_9.csv")
```

```python
df
```

```python
df.set_index('customer_id', inplace=True)
```

### **Dataset Visualization Using Plot() Function In Pandas Library:**

```python
df.plot(
   kind='barh',
   figsize=(16,4),
   color='purple',
   width=0.7,
   legend=False
)
plt.gca().invert_yaxis()
plt.title("Top 5 Customers According To Sum Of The Total Sales")
plt.xlabel("Total Sales In (USD$)")
plt.ylabel("Customer Id")
plt.tight_layout()
```
