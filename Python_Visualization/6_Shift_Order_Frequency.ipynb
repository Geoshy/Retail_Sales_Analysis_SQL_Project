### **Dataset Visualization Using Matplotlib Library:**

```python
plt.figure(figsize=(16,4))
plt.barh(
    df["shift_type"],
    df["order_count"],
    )
plt.gca().invert_yaxis()
plt.title("Relation Between Orders Count And Shift Time")
plt.xlabel("Orders Frequency")
plt.ylabel("Shift Type")
plt.tight_layout()
```
